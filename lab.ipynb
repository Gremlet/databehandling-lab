(e) Use the `boxplot()` method of `college` to produce side-by-side boxplots of `Outstate` versus `Private`.


````python
import matplotlib.pyplot as plt

college.boxplot(column="Outstate", by="Private")
plt.title("Out-of-state tuition vs Private/Public status")
plt.suptitle("")  # removes the automatic 'Boxplot grouped by ...' title
plt.xlabel("Private")
plt.ylabel("Outstate")
````
