1. One thing I noticed was that there was an odd value for one college in `Grad.Rate`. Cazenovia college has a graduation rate of 118, which is not possible, as it should be the proportion of students graduating within a certain time frame, so the maximum logical value is 100%


````python
college.loc[college["Grad.Rate"] > 100]
````

There are a few ways one could deal with this anomaly.

- Clipping: `college["Grad.Rate"] = college["Grad.Rate"].clip(upper=100)`
- Manual correction: setting that one row to 100 or the median.
- Removal: dropping it from analysis if it’s just a single outlier.

3. Do colleges with more applications also accept more students?


````python
college["Apps"].corr(college["Accept"])
````

- That is a very strong correlation with an r-value of ~0.9. Let's check if this is statistically significant by getting the p-value.


````python
from scipy.stats import pearsonr

r, p = pearsonr(college["Apps"], college["Accept"])
print("r:", r)
print("p:", p)
````

- The p-value is < 0.05 so the correlation is, indeed, significant i.e. the observed relationship is unlikely to have occurred by chance.


4. I wanted an overview of what other correlations might exist and thought a heatmap could give some nice visual insight.


````python
import seaborn as sns

# get only numeric columns, as we can't use corr with categorical columns like Elite
college_numeric = college.select_dtypes(include="number")

corr_matrix = college_numeric.corr()

sns.heatmap(corr_matrix, cmap="coolwarm")
````

- This shows some interesting correlations. Some are fairly obvious, like `Outstate` and `Expend` have a fairly strong positive correlation. We would expect that out-of-state tuition is higher at colleges that spend more per student. Some maybe a bit less obvious; like the modest negative correlation between `Room.Board` and `S.F.Ratio` meaning colleges with lower student-to-faculty ratios (i.e., more faculty per student) tend to charge more for room and board. To be expected, perhaps; but it's not a correlation I would have thought about off the top of my head. This shows that visualising correlations can be really useful; patterns you wouldn't necessarily think of can emerge.


### Aside

This might be beyond the scope of this lab but I found it very interesting how data invites human speculation. For instance, when I was looking into the odd 118% graduation rate outlier, I wondered which colleges have the lowest graduation rates and I noticed that Brigham Young University had a graduation rate of only 33%. That was surprising, as by all accounts BYU is a very decent university. That curiosity sent me looking for context, not for the sake of this lab, but because the number felt counterintuitive. It turns out that the explanation is largely cultural rather than academic. From a former BYU student ([source](https://www.quora.com/What-are-the-main-reasons-why-students-from-BYU-dont-graduate-on-time)):

> Without having actual statistics, and basing my answer on my observations there:
>
> 1. The mission puts you much older than typical students. Graduating on time at BYU means 24, not 22 years old.
> 2. Family commitments: Mormons are much more likely to marry young, and then have kids. I knew many BYU students with children. The mother often dropped out to raise the kids and work part time so the father could finish his studies.
> 3. Working: many Mormons don’t want a lot of student debt, so they work part time and take fewer courses. Former BYU President Rex Lee tried to fight this by telling students to graduate and pay off loans, but the temptation to work and pay as you go.
> 4. I hope this has changed, but many women went to BYU get married. They called that the MRS degree. They would drop out after marrying or having kids, or wait until the kids were older to finish. I believe BYU is still one of the few schools where men are more likely to graduate than women. Women outnumber men in college attendance, except in Utah Nationally, women are far outpacing men, and now hold more degrees despite decades of being behind.
> 5. Mormons do value education, but mostly for pragmatic reasons. They are encouraged to get high paying professions to raise their families and to help build up the church. They are not encouraged to go into the humanities or get educations purely for personal pleasure.

This made me realise something kind of fun about this whole process; exploring this kind of data can tell you _what_ is happening, but it also makes you wonder about the _why_ and this can lead you to sociology, culture, history, behaviour, psychology etc. All the complex and messy human stuff that doesn't fit neatly into a `.csv` I know that my little BYU dive isn't "data" in the formal sense. It was just me following a curiosity sparked by an odd number. But I love that numbers can push you to ask questions that the dataset can't fully answer.
